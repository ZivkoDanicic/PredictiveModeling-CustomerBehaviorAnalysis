# Statistical Testing - Association Between Qualitative Variables

```python
qualitative_columns = ['order_date', 'product_name', 'brand', 'color', 'subcategory', 'category', 'country', 'state', 'open_date', 'city', 'gender', 'age_group']
```

**Chi-Square Test**

```python
from scipy.stats import chi2_contingency


# Create contingency tables and calculate Chi-square values
chi_square_results = {}

for i in range(len(qualitative_columns)):
    for j in range(i + 1, len(qualitative_columns)):
        col1 = qualitative_columns[i]
        col2 = qualitative_columns[j]
        contingency_table = pd.crosstab(df[col1], df[col2])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        chi_square_results[(col1, col2)] = {
            'chi2': chi2,
            'p_value': p,
            'dof': dof,
            'expected': expected
        }

chi_square_results.keys()
```

We now have a dictionary call chi_square_results that contains the order pairs of our qualitative columns as keys, and the results of chi square stat and p-value contained in the items. We can now create a for loop to extract the results and check for statistical significance. The result will be lengthly, but we will be able to note all the associations between the qualitative columns and make any observations. 

```python
alpha = 0.05  # Significance level

for (col1, col2), results in chi_square_results.items():
    print(f"Chi-square results for {col1} vs {col2}:")
    print(f"  Chi-square value: {results['chi2']}")
    print(f"  p-value: {results['p_value']}")
    
    # Determine statistical significance
    if results['p_value'] <= alpha:
        print("  Result: Statistically significant")
    else:
        print("  Result: Not statistically significant")
    
    print()  # Print a blank line for better readability
```

Let's take the first result and put it in focus to interpret the result:

1. Chi-square Value:
Value: 4,168,302.08
The Chi-square statistic measures how expectations compare to actual observed data. A higher Chi-square value suggests a larger difference between observed and expected frequencies, indicating a strong association between the two variables.

2. p-value:
Value: 0.000...000167
The p-value indicates the probability of observing a Chi-square statistic as extreme as, or more extreme than, the one calculated if the null hypothesis were true (i.e., if there were no association between order_date and product_name). A very small p-value (typically less than 0.05) suggests that the observed relationship is unlikely to be due to chance.

3. Result: Statistically Significant
Because the p-value is significantly less than the common threshold of 0.05, we reject the null hypothesis. This suggests that there is a statistically significant association between order_date and product_name. In practical terms, this means that the distribution of products sold varies across different order dates in a way that is unlikely to be random.

4. Interpretation:
The results imply that certain products are more frequently ordered on specific dates. This could be influenced by various factors such as seasonality, marketing campaigns, holidays, or trends in consumer behavior.

**Summary of Chi-square Results**

**Strong Associations**: Variables like order_date, product_name, brand, and color appear to have strong relationships with many other qualitative variables in the dataset.

**Lack of Association**: Some variables, particularly certain combinations (like brand vs country), do not demonstrate a significant relationship.

**Statistical Implication**s: The results imply that in your dataset, certain attributes (like order_date and product_name) influence other characteristics significantly, which could inform decisions or further analyses, while others may not be relevant for your investigation.

**Cramer's V** 

Add the Cramer's V coefficient into our code form above:

```python


# Initialize a dictionary to store Chi-square results
chi_square_results = {}

# Loop through unique pairs of qualitative columns
for i in range(len(qualitative_columns)):
    for j in range(i + 1, len(qualitative_columns)):
        col1 = qualitative_columns[i]
        col2 = qualitative_columns[j]
        contingency_table = pd.crosstab(df[col1], df[col2])
        
        # Perform the Chi-square test
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        
        # Calculate Cramér's V
        n = contingency_table.sum().sum()  # Total number of observations
        min_dim = min(contingency_table.shape)  # Minimum dimension (rows or columns)
        cramer_v = np.sqrt(chi2 / (n * (min_dim - 1)))

        # Store the results in the dictionary
        chi_square_results[(col1, col2)] = {
            'chi2': chi2,
            'p_value': p,
            'dof': dof,
            'expected': expected,
            'cramer_v': cramer_v
        }

alpha = 0.05  # Significance level

# Output the results
for (col1, col2), results in chi_square_results.items():
    print(f"Chi-square results for {col1} vs {col2}:")
    print(f"  Chi-square value: {results['chi2']}")
    print(f"  p-value: {results['p_value']}")
    print(f"  Cramér's V: {results['cramer_v']:.4f}")  # Display Cramér's V with 4 decimal places
    
    # Determine statistical significance
    if results['p_value'] <= alpha:
        print("  Result: Statistically significant")
    else:
        print("  Result: Not statistically significant")
    
    print()  # Print a blank line for better readability
```

# # Statistical Testing - Association Between Quantitative and Qualitative Variables

```python
# Import ANOVA function from SciPy
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Conduct ANOVA
anova_result = ols('profit ~ C(category)', data=df).fit()
anova_table = sm.stats.anova_lm(anova_result, typ=2)

# Check if ANOVA is significant
if anova_table['PR(>F)'][0] < 0.05:
    print("ANOVA is significant, performing Tukey's HSD.")
    
    # Perform Tukey's HSD test
    tukey_result = pairwise_tukeyhsd(endog=df['profit'], groups=df['category'], alpha=0.05)
    print(tukey_result)
else:
    print("ANOVA is not significant; no post-hoc test needed.")
```

**ANOVA Results**

Significance: Since ANOVA was significant (p < 0.05), it suggests that there are differences in profit among at least some of the product categories.

**Tukey's HSD Results**

The Tukey’s HSD test results indicate which **specific group differences are significant**, providing p-values for each comparison.

1. group1 and group2: These columns indicate the two categories being compared.

2. meandiff: This is the mean difference in profit between the two groups. A positive value indicates that the first group (group1) has a higher average profit than the second group (group2).

3. p-adj: This is the adjusted p-value for the comparison. A p-adj value less than 0.05 indicates a statistically significant difference between the two groups.

4. lower and upper: These columns show the confidence interval for the mean difference. If the interval does not include zero, it supports the conclusion of a significant difference.

5. reject: A value of True means that the null hypothesis (no difference in means) is rejected for that comparison.

**Summary of Key Comparisons** 

- Audio vs. Other Categories: The category "Audio" shows significant differences in profit compared to all other categories except for "Music, Movies and Audio Books," where the mean difference is not significant (p-adj = 0.288). This suggests that while "Audio" generally performs better than many categories, it does not differ significantly from "Music, Movies and Audio Books."

- Cameras and Camcorders: This category has a significant negative mean difference compared to several categories like "Audio," "Games and Toys," and "Home Appliances," indicating it earns significantly less profit than these categories.

- Games and Toys vs. Home Appliances: "Games and Toys" is significantly less profitable than "Home Appliances" but more profitable than "Cameras and Camcorders."

- Home Appliances: This category shows a high mean difference with "Games and Toys" and "Music, Movies and Audio Books," indicating it has a significantly higher average profit compared to these.

**Overall Insights**

- Highly Profitable Categories: "Audio," "Home Appliances," and "Computers" appear to be the most profitable, as they consistently show higher mean profits compared to other categories.


- Underperforming Categories: "Cameras and Camcorders" and "Games and Toys" show lower profits relative to several other categories, which may warrant further investigation or strategic adjustments.

```python
def categoricalplots(columns, figsize=(6, 4)):
    figures = []
    
    for col in columns:
        # Sum the profits for each category
        profit_by_category = df.groupby(col)['profit'].sum().reset_index()
        
        # Create a new figure for each plot
        fig = plt.figure(figsize=figsize)
        sns.barplot(x=col, y='profit', data=profit_by_category, alpha=0.5)
        plt.xlabel(col)
        plt.ylabel('Total Profit')
        plt.title(f'Total Profit by {col}')
        plt.xticks(rotation=90)  
        figures.append(fig)  # Store the figure
    
    for fig in figures:
        display(fig)  # Display all figures at once
```
